```python
import numpy as np

# 假设你已经定义了 baseball 列表
baseball = [[180, 78.4],
            [215, 102.7],
            [210, 98.5],
            [188, 75.2]]

np_baseball = np.array(baseball)

# 打印出 np_baseball 的加法结果，假设每个元素都增加 5
updated = np_baseball + 5
print("Updated np_baseball after addition:\n", updated)

# 创建转换数组，假设第1列是把英寸转换成米(1英寸=0.0254米)，第2列是把磅转换成千克(1磅=0.453592)
conversion = np.array([0.0254, 0.453592])

# 打印出 np_baseball 和 conversion 的逐元素乘积
converted_baseball = np_baseball * conversion
print("Converted np_baseball:\n", converted_baseball)


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate the secant values, avoiding division by zero
sec_x = 1 / np.cos(x)

# Set the limit to avoid plotting at points where cos(x) = 0
sec_x[np.abs(np.cos(x)) < 0.01] = np.nan  # Setting to NaN to avoid plotting vertical asymptotes

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, sec_x, label='sec(x)', color='blue')

# Add vertical lines for the asymptotes (where cos(x) = 0)
asymptotes_x = np.arange(-2 * np.pi, 2 * np.pi, np.pi/2)
plt.vlines(asymptotes_x, -10, 10, colors='red', linestyles='dashed', label='Asymptotes')

# Customize the plot
plt.title('Secant Function: sec(x)')
plt.xlabel('x (radians)')
plt.ylabel('sec(x)')
plt.ylim(-10, 10)  # Limit y-axis for better visualization
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
